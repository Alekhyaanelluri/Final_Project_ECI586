11 - month: last contact month of year (categorical: "jan", "feb", "mar", …, "nov", "dec")
12 - duration: last contact duration, in seconds (numeric)

13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)
15 - previous: number of contacts performed before this campaign and for this client (numeric)
16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

Output variable (desired target):
17 - y - has the client subscribed a term deposit? (binary: "yes","no")

Reflects how many predicted yes subscriptions were correct. A high precision score suggests the model avoids false positives.
Recall:
Measures how many actual yes subscriptions were correctly predicted. Low recall indicates the model misses many potential subscribers.
F1-Score:
Balances precision and recall, providing a single metric to evaluate model effectiveness in imbalanced datasets.

## K Nearest Neighbors

**the steps for making this knn model:**

1. creating a copy of the dataset.
2. converting all X variables that are factors to numeric
3. scaling all the X variables.
4. using linear regression for each x to y for feature selection
5. using accuracy to decide the best k for the model.
6. making a confution matrix and calculating classification metrics for each step (for comparison).

### preprocessing

```r
knn.df <- Mt_df
head(knn.df)
```

```r
for (i in 1:16)
{
    knn.df[,i] <- as.numeric(knn.df[,i])
}
knn.df$pdays[knn.df$pdays==-1]=0 
knn.df$previous[knn.df$previous==-1]=0 
head(knn.df)
```

```r
knn.df1 <- sapply(knn.df[,1:16],scale)
knn.df1 <- as.data.frame(knn.df1)
knn.df <- cbind(knn.df1,knn.df$y)
head(knn.df)
```

### feature selection

```r
# significance level using linear regression
knn.df1 <- knn.df 
knn.df1[,17] <- as.numeric(knn.df1[,17])
for (i in 1:16)
{
  print(names(knn.df1)[i]) 
  print(summary(lm(knn.df1[,17]~knn.df1[,i],knn.df1)))
}
```

```r
set.seed(8)
train.size = floor(0.75*nrow(knn.df))
train.index = sample(1:nrow(knn.df), train.size)
train.set = knn.df[train.index,]
test.set = knn.df[-train.index,]

x.train = train.set[,-17] 
x.test = test.set[,-17] 
y.train = train.set[,17] 
y.test = test.set[,17] 

knn.3 <- knn(train = x.train, test = x.test, cl = y.train , k = 3)
TB = table(predicted = knn.3, true = y.test)
accuracy = round((TB[1]+TB[4])/sum(TB)*100,2)
accuracy
```

```r
TB
```

```r
library(caret)

# data/code from "2 class example" example courtesy of ?caret::confusionMatrix

lvs <- c("no", "yes")
truth <- factor(rep(lvs, times = c(9940, 1363)),
                levels = rev(lvs))
pred <- factor(
  c(
    rep(lvs, times = c(9539, 401)),
    rep(lvs, times = c(853, 510))),
  levels = rev(lvs))

caret::confusionMatrix(pred, truth)
```

```r
table <- data.frame(caret::confusionMatrix(pred, truth)$table)

plotTable <- table %>%
  mutate(goodbad = ifelse(table$Prediction == table$Reference, "good", "bad")) %>%
  group_by(Reference) %>%
  mutate(prop = Freq/sum(Freq))

# fill alpha relative to sensitivity/specificity by proportional outcomes within reference groups (see dplyr code above as well as original confusion matrix for comparison)
ggplot(data = plotTable, mapping = aes(x = Reference, y = Prediction, fill = goodbad, alpha = prop)) +
  geom_tile() +
  geom_text(aes(label = Freq), vjust = .5, fontface  = "bold", alpha = 1,size = 30) +
  scale_fill_manual(values = c(good = "forestgreen", bad = "firebrick1")) +
  theme_bw() +
  xlim(rev(levels(table$Reference))) +
theme(text = element_text(size=30)) +
xlab("True Value")
```

```r
fig(15,10)
k = seq(1,29,2)
i=1 
Accuracy=1                     

for (i in k)
{
  knn.mod <-  knn(train=x.train, test=x.test, cl=y.train, k=i)
  Accuracy[i] <- 100 * sum(y.test == knn.mod)/length(y.test)
  k=i  
  cat(k,'=',Accuracy[i],'\n')        
}

Accuracy = Accuracy[!is.na(Accuracy)]
k = seq(1,29,2)
Accuracyplot = data.frame(k,Accuracy)
#plot our k
ggplot(Accuracyplot,aes(x=k,y=Accuracy))+
  geom_line(col = "red")+
  expand_limits(y=Accuracy[0])+
  theme_dark() +
  geom_text(aes(label=round(Accuracy,1)),vjust = -0.5,size = 10) + 
  theme(text = element_text(size=30))

```

# Key Findings and Insights
The Random Forest model that performed the best at the train data has performed extremely well at the test data, the main aim here is to maximize the bank's profit by idetifying as much people that will use the bank's term deposit,
The model succesfully identifyed **84%** of the poeple that chose to use this deposit (**438 out of 521**). cohen's kappa statistic has shown a moderate agreement between the prection and the actual data which is a good indicator of the reliebility of the model.

Best Performing Model:
The Random Forest model emerged as the most effective tool for predicting term deposit subscriptions, demonstrating both high accuracy and reliability in identifying potential customers.
On the test data, the model successfully identified 84% of customers who subscribed to the deposit (438 out of 521), showcasing its practical utility.

Cohen's Kappa Statistic:
The model achieved a moderate agreement between predictions and actual outcomes, as indicated by Cohen's Kappa statistic. This underscores the reliability and consistency of the model in handling the imbalanced nature of the dataset.

Feature Importance:
Key predictors, such as duration of the call, pdays (days since the last contact), and campaign (number of contacts), significantly influenced the likelihood of subscription.
Longer call durations were strongly associated with positive outcomes, emphasizing the importance of meaningful customer interactions.

Class Imbalance Challenges:
While the model performed well in identifying "yes" subscriptions, the imbalanced nature of the dataset (with a majority of "no" responses) posed challenges for improving recall on the minority class.
Addressing this imbalance through advanced techniques, such as SMOTE or cost-sensitive learning, could further enhance the model's effectiveness.

Predictive Power of Campaign Metrics
Variables related to campaign engagement, such as the duration of the call and number of contacts, were among the most significant predictors of whether a customer would subscribe to a term deposit.
Longer calls often indicated greater customer interest, suggesting that meaningful interactions during telephonic campaigns are crucial. Additionally, timely follow-ups positively influenced subscription rates.

Impact of Demographics
Demographic variables like age, job, and education showed noticeable patterns. For instance:
Customers in certain professional categories (e.g., "management" or "self-employed") had higher subscription rates.
Older customers (above 40) were more likely to subscribe compared to younger age groups.
These patterns suggest that targeting specific demographic segments could improve campaign efficiency.

Role of Economic Indicators
Macroeconomic variables like the employment variation rate and consumer confidence index had moderate predictive power.
Economic stability positively influenced subscription likelihood, aligning with the idea that customers are more willing to commit to financial products during favorable economic conditions.

Performance of Predictive Models
Logistic regression and random forest models were evaluated for their ability to classify subscription outcomes:
Logistic regression provided interpretable coefficients for understanding feature importance but struggled with complex interactions.
Random forest achieved better accuracy and feature importance rankings but required careful tuning to avoid overfitting.

Challenges and Limitations
Despite strong performance on the majority class, the models struggled with recall for the minority class (y = yes), leading to missed opportunities in identifying potential subscribers.
Addressing these challenges requires additional data (e.g., psychometric or behavioral information) or advanced techniques like ensemble learning.

Actionable Recommendations
Optimize Call Strategies: Focus on meaningful, longer interactions during calls.
Target Specific Segments: Prioritize campaigns for older customers and certain professional groups with higher subscription likelihoods.
Refine Follow-Up Timing: Use insights from pdays to schedule optimal follow-ups for maximizing conversions.
Improve Model Performance: Address class imbalance through resampling or cost-sensitive learning methods.

# Limitations and Ethical/Legal Considerations
Limitations of the Analysis

Feature Limitations:
The dataset may lack critical predictors, such as customer behavior data (e.g., previous banking activity) or psychometric measures.
Impact: The absence of these features might limit the model’s predictive power.

Model Interpretability:
While simpler models like logistic regression are interpretable, complex models (e.g., random forests) can be harder to explain to non-technical stakeholders.
Impact: This may hinder adoption and trust in the models.

Time-Specific Analysis:
The data reflects a specific time period, and external factors (e.g., economic conditions) might have shifted since then.
Impact: Predictions may not hold true for future campaigns without retraining the models.

Ethical and Legal Considerations

Customer Privacy:
The data includes sensitive demographic and behavioral information about customers.
Action Taken: Ensure that any analysis and sharing of insights comply with data privacy regulations such as GDPR, anonymizing data where necessary.

Bias and Fairness:
Models trained on demographic features like age, marital status, or education might unintentionally perpetuate biases.
Action Taken: Evaluate model predictions to ensure they are not unfairly biased against specific groups.

Transparency and Accountability:
Predictive models, especially black-box ones, may lead to decisions that are difficult to justify to stakeholders or customers.
Action Taken: Use interpretable metrics and visualizations to explain model outcomes and ensure transparency.

Informed Consent:
Customers may not have explicitly consented to their data being used for predictive modeling.
Action Taken: Acknowledge the importance of ensuring ethical data collection practices that respect customer consent.
